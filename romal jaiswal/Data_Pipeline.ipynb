# Model Training & testing

```python
from sklearn.model_selection import train_test_split


x = df.drop('MY rate', axis = 1)
y = df['MY rate']


x_input = x
```

```python
x = pd.get_dummies(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor




lr = LinearRegression()
knnr = KNeighborsRegressor()
dtr = DecisionTreeRegressor(random_state=0)
xgbr = XGBRegressor()
gbr = GradientBoostingRegressor(random_state=0)
abr = AdaBoostRegressor(random_state=0, n_estimators=100)
rfr = RandomForestRegressor(max_depth=2, random_state=0)
```

```python
algo_list = [lr , knnr , dtr , xgbr , gbr , abr , rfr]
```

```python
perf = pd.DataFrame()

algos = []
train_scores = []
test_scores = []

for algo in algo_list:
  model = algo.fit(x_train , y_train)
  train_score = model.score(x_train, y_train)
  test_score = model.score(x_test, y_test)

  algos.append(algo)
  train_scores.append(train_score)
  test_scores.append(test_score)

perf['algo'] = algos
perf['train_score'] = train_scores
perf['test_score'] = test_scores


perf.sort_values(by=['test_score'])
```
